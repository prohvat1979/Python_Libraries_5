1. id: Represents a unique ID of each patient — (представляет уникальный идентификатор каждого пациента.

```python
# Смотрим количество пропущеных значений.
print("Количество пропущенных значений:")
print(df.isnull().sum())
```

```python
# Удаляем колонку с нулевыми значениями.
df_cancer = df.drop('Unnamed: 32', axis=1)
```

```python
# Смотрим тип данных
df_cancer.dtypes
```

```python
# Статистический анализ числовых переменных
df_cancer.describe().T
```

```python
# Визуализация гистограмм
plt.figure(figsize=(40, 40))
for i, column in enumerate(['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
                            	'area_mean', 'smoothness_mean', 'compactness_mean',	'concavity_mean',
                                'concave points_mean', 'texture_worst',	'perimeter_worst',	'area_worst',
                            	'smoothness_worst',	'compactness_worst', 'concavity_worst', 'concave points_worst',
                            	'symmetry_worst',	'fractal_dimension_worst'], 1):
    plt.subplot(8, 3, i)
    sns.histplot(df[column], kde=True)

plt.tight_layout()
plt.show()
```

```python
# Подготавливаем столбец diagnosis перед обученим модели
# Приводим обектные значения к интовым
mapDiag = {'B': 0, 'M': 1}
df_cancer['diagnosis'] = df_cancer['diagnosis'].map(mapDiag)
```

```python
# Проверяем тип чтобы удостовериться, что изменения прошли успешно
df_cancer.dtypes['diagnosis']
```

```python
# Извлечём diagnosis из признаков
X = df_cancer.drop('diagnosis', axis=1)
# Определим целевую переменную
y = df_cancer['diagnosis']
```

```python
# Определяем переменные для обучения
#X = df_features
#y = target

# Маштабируем признаки
X_scaled = StandardScaler().fit_transform(X)
# Разделяем данные на обучающий и тестовые наборы
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, train_size=0.8, random_state = 42)

# Выводим размеры
print("Размер обучающего набора:", X_train.shape)
print("Размер тестового набора:", X_test.shape)
```

```python
# Разные значения C (обратные значения скорости обучения)
C_values = [0.001, 0.01, 0.1, 1, 10, 100]
results = []

for C in C_values:
    # Создание и обучение модели
    model = LogisticRegression(C=C, solver='liblinear', max_iter=1000)
    start_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - start_time

    # Предсказание на тестовой выборке
    y_pred = model.predict(X_test)

    # Оценка качества модели
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    results.append((C, accuracy, precision, recall, f1, training_time))
```

```python
# Вывод результатов
for result in results:
    print(f'C: {result[0]}, \
        \n Accuracy: {result[1]}, \
        \n Precision: {result[2]}, \
        \n Recall: {result[3]}, \
        \n F1 Score: {result[4]}, \
        \n Training Time: {result[5]:.4f} sec')
```

**2. Познакомиться с работой стохастического градиентного спуска.**
* Используйте набор данных для обучения модели с использованием стохастического градиентного спуска (SGD).
* Сравните скорость сходимости и итоговое качество модели со стандартным градиентным спуском.

```python
# Создание и обучение модели с использованием стандартного градиентного спуска
start_time = time.time()
model_lr = LogisticRegression(class_weight='balanced')
model_lr.fit(X_train, y_train)
training_time_lr = time.time() - start_time

# Предсказание на тестовой выборке
predictions_lr = model_lr.predict(X_test)

# Оценка производительности модели
accuracy_lr = accuracy_score(y_test, predictions_lr)
print("Accuracy LogisticRegretion:", accuracy_lr)
```

```python
# Создание и обучение модели с использованием стохастического градиентного спука (SGD)
start_time = time.time()
model_sgd = SGDClassifier(class_weight='balanced')
model_sgd.fit(X_train, y_train)
training_time_sgd = time.time() - start_time

# Предсказание на тестовой выборке
predictions_sgd = model_sgd.predict(X_test)

# Оценка производительности модели
accuracy_sgd = accuracy_score(y_test, predictions_sgd)
print("Accuracy SGDClassifier:", accuracy_sgd)
```

```python
print(f'Logistic Regression: \
        \n Accuracy: {accuracy_lr} \
        \n Training Time: {training_time_lr:.4f} sec')

print(f'SGDClassifier: \
        \n Accuracy: {accuracy_sgd} \
        \n Training Time: {training_time_sgd:.4f} sec')
```

Модель логистической регрессии показала более высокую точность по сравнению с SGD-классификатором, но SGD-классификатор значительно быстрее обучается, чем логистическая регрессия.

**3. Сравнить разные алгоритмы оптимизации.**
* На исходном наборе данных обучить модель логистической регрессии, используя разные оптимизаторы, и сравнить их эффективность.

```python
# Алгоритмы оптимизации для LogisticRegression
solvers_lr = ['liblinear', 'lbfgs', 'newton-cg', 'sag', 'saga']
results_optimizers_lr = []

for solver_lr in solvers_lr:
    start_time = time.time()
    model_opt_lr = LogisticRegression(solver=solver_lr, max_iter=1000)
    model_opt_lr.fit(X_train, y_train)
    training_time_opt_lr = time.time() - start_time

    y_pred = model_opt_lr.predict(X_test)
    accuracy_opt_lr = accuracy_score(y_test, y_pred)

    results_optimizers_lr.append((solver_lr, accuracy_opt_lr, training_time_opt_lr))

# Вывод результатов
print('All Logistic Regression:')
for result_opt_lr in results_optimizers_lr:
    print(f'\n Algoritm: {result_opt_lr[0]}, \
        \n Accuracy: {result_opt_lr[1]}, \
        \n Training Time: {result_opt_lr[2]:.4f} sec')
```

Алгоритмы показывают одинаковую точность, это говорит о том, что все они хорошо справляются однако есть разница в времени выполнения, liblinear имеет максимальную скорость выполнения в 0.0071 секунду, а saga дольше всего справляется с данной задачей с временем в 0.125 секунды.

```python
# Алгоритмы оптимизации для SGDClassifier
lossed_sgd = ['log_loss', 'squared_epsilon_insensitive', 'perceptron', 'modified_huber', 'hinge', 'epsilon_insensitive', 'huber', 'squared_error', 'squared_hinge']
results_optimizers_sgd = []

for loss_sgd in lossed_sgd:
    start_time = time.time()
    model_opt_sgd = SGDClassifier(loss=loss_sgd, max_iter=1000)
    model_opt_sgd.fit(X_train, y_train)
    training_time_opt_sgd = time.time() - start_time

    y_pred = model_opt_sgd.predict(X_test)
    accuracy_opt_sgd = accuracy_score(y_test, y_pred)

    results_optimizers_sgd.append((loss_sgd, accuracy_opt_sgd, training_time_opt_sgd))

# Вывод результатов
print('All SGDClassifier:')
for result_opt_sgd in results_optimizers_sgd:
    print(f'\n Algoritm: {result_opt_sgd[0]}, \
        \n Accuracy: {result_opt_sgd[1]}, \
        \n Training Time: {result_opt_sgd[2]:.4f} sec')
```

Лучше всего справился с датасетом hinge с счётом 0.9824561403508771 а так же является самым быстрым 0.0040 секунд, что делает его лучшим выбором для данной задачи.

Хуже всего справился squared_error с счётом 0.4298245614035088 он однозначно не подходит для данной задачи.

```python
# Визуализация результатов
optimizer_sgd_df = pd.DataFrame(results_optimizers_sgd, columns=["Loss", "Accuracy", "Training Time"])
optimizer_sgd_df.set_index("Loss")[["Accuracy", "Training Time"]].plot(
    kind="bar", figsize=(12, 6), secondary_y="Training Time", color=["skyblue", "orange"]
)
plt.title("Сравнение методов оптимизации SGDClassifier")
plt.ylabel("Accuracy")
plt.grid(axis="y")
plt.show()
```
